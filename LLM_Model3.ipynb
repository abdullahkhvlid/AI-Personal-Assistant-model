```python
from gpt4all import GPT4All

model = GPT4All("mistral-7b-instruct-v0.1.Q4_0.gguf")

```

```python
def ask(prompt):
    return model.generate(prompt, max_tokens=600)

name = input("Enter your name: ")
g1 = input("Goal 1: ")
g2 = input("Goal 2: ")
g3 = input("Goal 3: ")
hours = input("Study hours: ")
topic = input("Focus topic: ")

full_prompt = f"""
Create a personalized productivity plan for a student named {name}.
Goals: {g1}, {g2}, {g3}.
Available study hours: {hours}.
Focus topic: {topic}.
Include time blocks, study suggestions, break ideas, and task priorities.
"""

plan = ask(full_prompt)

summary_prompt = f"""
Summarize the following plan in exactly 5 bullet points.
Plan: {plan}
"""

summary = ask(summary_prompt)

motivation_prompt = "Give 3 short motivational lines for a student starting their daily routine."

motivation = ask(motivation_prompt)

print("\nPersonalized Schedule\n")
print(plan)

print("\nSummary\n")
print(summary)

print("\nMotivation\n")
print(motivation)


```

```python
tip_choice = input("Do you want a productivity tip? (yes/no): ")

if tip_choice.lower() == "yes":
    tip = ask("Give one practical productivity tip for students.")
    print("\nProductivity Tip\n")
    print(tip)
```
